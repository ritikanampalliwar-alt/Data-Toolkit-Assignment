# Code Questions & Answers

```python
# 1. How do you create a 2D NumPy array and calculate the sum of each row?
# Answer:
# This program creates a 2D NumPy array and calculates the sum of elements in each row.
# Example:
import numpy as np

arr = np.array([[1, 2, 3],
                [4, 5, 6]])
row_sum = arr.sum(axis=1)
print(row_sum)
```

```python
# 2. Write a Pandas script to find the mean of a specific column in a DataFrame.
# Answer:
# This program calculates the average value of a selected column.
# Example:
import pandas as pd

df = pd.DataFrame({'Marks': [70, 80, 90]})
print(df['Marks'].mean())
```

```python
# 3. Create a scatter plot using Matplotlib.
# Answer:
# This program creates a simple scatter plot.
# Example:
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.scatter(x, y)
plt.show()
```

```python
# 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
# Answer:
# This program calculates correlation and displays it using a heatmap.
# Example:
import seaborn as sns
import pandas as pd

data = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

sns.heatmap(data.corr(), annot=True)
```

```python
# 5. Generate a bar plot using Plotly.
# Answer:
# This program generates an interactive bar chart.
# Example:
import plotly.express as px

data = {'Names': ['A', 'B', 'C'], 'Values': [10, 20, 30]}
fig = px.bar(data, x='Names', y='Values')
fig.show()
```

```python
# 6. Create a DataFrame and add a new column based on an existing column.
# Answer:
# This program adds a new column derived from another column.
# Example:
import pandas as pd

df = pd.DataFrame({'Salary': [20000, 30000]})
df['Bonus'] = df['Salary'] * 0.1
print(df)
```

```python
# 7. Write a program to perform element-wise multiplication of two NumPy arrays.
# Answer:
# This program multiplies two arrays element by element.
# Example:
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(a * b)
```

```python
# 8. Create a line plot with multiple lines using Matplotlib.
# Answer:
# This program plots two lines on the same graph.
# Example:
import matplotlib.pyplot as plt

x = [1, 2, 3]
y1 = [2, 4, 6]
y2 = [1, 3, 5]

plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

```python
# 9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.
# Answer:
# This program filters data based on a condition.
# Example:
import pandas as pd

df = pd.DataFrame({'Score': [45, 75, 90]})
print(df[df['Score'] > 60])
```

```python
# 10. Create a histogram using Seaborn to visualize a distribution.
# Answer:
# This program displays data distribution using a histogram.
# Example:
import seaborn as sns

data = [10, 20, 20, 30, 40]
sns.histplot(data)
```

```python
# 11. Perform matrix multiplication using NumPy.
# Answer:
# This program multiplies two matrices using NumPy.
# Example:
import numpy as np

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
print(np.dot(a, b))
```
